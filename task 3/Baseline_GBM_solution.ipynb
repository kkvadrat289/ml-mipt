```python
consts = []
for Id in my_train['item_id']:
    matrix1 = my_train[(my_train['item_id']==Id)&(my_train['shift']==1)].as_matrix()
    if len(matrix1) > 1:
        summ=0
        for i in range(0, len(matrix1)-1):
            summ += matrix1[i][1]/matrix1[i+1][-1]
        consts.append(summ/(len(matrix1)-1))
```

```python
summ = 0
n = 0
for item in consts:
    if item < 10:
        summ += item
        n +=1
Const = summ/n
print Const
```
